```python
import pickle
import pandas as pd
```

```python
embeddings, labels = pickle.load( open( "embeddings.p", "rb" ) )
```

```python
emb = []
lab = []
for i, label in enumerate(labels):
        emb.append(embeddings[i].tolist())
        lab.append(label)
tokens_df = pd.DataFrame(emb)
labels_df = pd.DataFrame(lab)
tokens_df.to_csv('rnn_emb.csv', sep='\t', index=False, header=None)
labels_df.to_csv('rnn_lab.csv',sep='\t', index=False, header=None)
```
